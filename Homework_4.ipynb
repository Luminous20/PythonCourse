```

import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tt = pd.read_csv('https://raw.githubusercontent.com/carlson9/KocPython2020/master/homework/immSurvey.csv')
tt.head()

```

```
alphas = tt.stanMeansNewSysPooled
sample = tt.textToSend
```

```
bigram_vectorizer = CountVectorizer(ngram_range=(1, 2),
...                                     token_pattern=r'\b\w+\b', min_df=1)
```

```
bianalyse = bigram_vectorizer.build_analyzer()
```

```
for line in sample:
      bianalyse(line)
```

```
bianalyse("".join(sample))
```

```
X_2 = bigram_vectorizer.fit_transform(sample).toarray()
```

```
X_2.max()
```

```
pd.DataFrame(X_2)
```

```

from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_2, alphas,
random_state=1)

```

```

from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import ConstantKernel, RBF

```

```

rbf = ConstantKernel(1.0) * RBF(length_scale=1.0)
gpr = GaussianProcessRegressor(kernel=rbf, alpha=1e-8)
```

```

gpr.fit(Xtrain, ytrain)

```

```

# Compute posterior predictive mean and covariance
mu_s, cov_s = gpr.predict(Xtest, return_cov=True)

```

```

#test correlation between test and mus
np.corrcoef(ytest, mu_s)
```
